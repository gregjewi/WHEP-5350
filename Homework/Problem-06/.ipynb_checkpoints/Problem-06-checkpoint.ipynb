![](ScheduleB.png)


**C) Indicate the dates when the reservoir was operating in Condition C.**

The minimum reservoir elevation and outflow during 2008 were 681.87 feet NGVD 580 cfs respectively. These values suggest that there were zero days in which the reservoir was operating under any condition C. 

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
LakeLevel = pd.read_csv(
    'CoralvilleLakeDailies2008.txt',
    sep = '\t',
    names = ['Agency','Site Number','Date','Lake Elevation','Approved'],
    skiprows = 29,
    usecols = [2,3]
)

LakeLevel.Date = pd.to_datetime(LakeLevel.Date)
LakeLevel = LakeLevel.set_index('Date');
```

```python
Flow = pd.read_csv(
    'IowaRiverAtCoralvilleLakeDailies.txt',
    sep = '\t',
    names = ['Agency','Site Number','Date','Flow','Approved'],
    skiprows = 29,
    usecols = [2,3]
)

Flow.Date = pd.to_datetime(Flow.Date);
Flow = Flow.set_index('Date');
```

```python
Data = pd.concat([LakeLevel,Flow],axis=1)

Month = pd.concat([Flow[Flow.index.day == 1], LakeLevel[LakeLevel.index.day == 1]],axis=1)
Condition = ['Nan','Nan','A-2','A-2','B-1','B-1','B-3','B-3','A-4','Nan','Nan','Nan']
Month['Condition'] = pd.Series(Condition,index=Month.index)

MonthShow = Month['20080301':'20080901']
# print(tabulate(MonthShow,headers=MonthShow.columns.values,tablefmt='pipe'))
```

```python
ConservativePool = pd.Series(index=Data.index)
ConservativePool['20080101':'20080214'] = 683
ConservativePool['20080215':'20080320'] = 683
ConservativePool['20080321':'20080520'] = 679
ConservativePool['20080521':'20080915'] = 683
ConservativePool['20080915':'20081215'] = 686
ConservativePool['20081216':'20081231'] = 683

Data['Conservative Pool'] = ConservativePool

Data['Spillway'] = pd.Series(712,index=Data.index)
Data['Flood Operations'] = pd.Series(707,index=Data.index)
```

```python
fig,ax = plt.subplots(figsize=(12,10));

Data['Flow'].plot(ax=ax,legend=True,secondary_y=True,xlim=['20080101','20081231']);
Data['Lake Elevation'].plot(ax=ax,legend=True);
Data['Conservative Pool'].plot(ax=ax,legend=True,linestyle='--',color='gray')
Data['Spillway'].plot(ax=ax,legend=True,linestyle='-.',color='gray')
Data['Flood Operations'].plot(ax=ax,legend=True,linestyle=':',color='gray')


plt.title('2008 Reservoir Pool and Outflow of Coralville Lake');

ax.grid(True, axis='x',linestyle='--');
plt.savefig('Full2008.png');
```

```python
fig,ax = plt.subplots(figsize=(12,10));
Data[Data['Lake Elevation'] >= 707]['Flow'].plot(ax=ax,legend=True);
Data[Data['Lake Elevation'] >= 707]['Lake Elevation'].plot(ax=ax,secondary_y=True,legend=True);

plt.title('Days in 2008 where Schedule B-1, B-2, B-3 \n Pool Elevation > 707');

ax.grid(True, axis='x',linestyle='--');
plt.savefig('ScheduleB.png');
```

```python
print('Minimum Reservoir Elevation: ', Data['Lake Elevation'].min())
print('Minimum Flow: ', Data.Flow.min())
```
